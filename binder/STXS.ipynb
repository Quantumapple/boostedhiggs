```python
import scipy
def get_finetuned_score(data, model_path):
    import onnx
    import onnxruntime as ort

    input_dict = {
        "highlevel": data.loc[:, "fj_ParT_hidNeuron000":"fj_ParT_hidNeuron127"].values.astype("float32"),
    }

    onnx_model = onnx.load(model_path)
    onnx.checker.check_model(onnx_model)

    ort_sess = ort.InferenceSession(
        model_path,
        providers=["AzureExecutionProvider"],
    )
    outputs = ort_sess.run(None, input_dict)

    return scipy.special.softmax(outputs[0], axis=1)[:, 0]
```

```python
with open("../fileset/luminosity.json") as f:
    luminosity = json.load(f)["ele"]["2017"]

def get_sum_sumgenweight(pkl_files, year, sample):
    """Load and sum the sumgenweight of each pkl file."""

    sum_sumgenweight = 0
    for ifile in pkl_files:
        with open(ifile, "rb") as f:
            metadata = pkl.load(f)
        sum_sumgenweight = sum_sumgenweight + metadata[sample][year]["sumgenweight"]

    return sum_sumgenweight

def get_xsecweight(pkl_files, year, sample, is_data, luminosity):
    """Get xsec-weight and scales events by lumi/sumgenweights."""

    if not is_data:
        # find xsection
        f = open("../fileset/xsec_pfnano.json")
        xsec = json.load(f)
        f.close()
        try:
            xsec = eval(str((xsec[sample])))
        except ValueError:
            print(f"sample {sample} doesn't have xsecs defined in xsec_pfnano.json so will skip it")
            return None

        # get overall weighting of events.. each event has a genweight...
        # sumgenweight sums over events in a chunk... sum_sumgenweight sums over chunks
        xsec_weight = (xsec * luminosity) / get_sum_sumgenweight(pkl_files, year, sample)
    else:
        xsec_weight = 1
    return xsec_weight
```

```python
channels = [
    "ele",
    "mu",
]

samples = [
    "ggF",
    "VBF",
]

samples_dict = {
    "ggF": "GluGluHToWW_Pt-200ToInf_M-125_Rivet",
    "VBF": "VBFHToWWToAny_M-125_TuneCP5_withDipoleRecoil_Rivet",
}

samples_dir = "../eos/Dec20_hww"
years = [
    "2016",
    "2016APV",
    "2017",
    "2018"
]
```

```python
df = {}

for sample in samples:
    df[sample] = {}
    for year in years:
        df[sample][year] = {}
                   
        for ch in channels:

            df[sample][year][ch] = pd.read_parquet(glob.glob(f"{samples_dir}_{year}/{samples_dict[sample]}/outfiles/*_{ch}.parquet"))

            df[sample][year][ch]["THWW"] = get_finetuned_score(df[sample][year][ch], "../../weaver-core-dev/experiments_finetuning/v35_30/model.onnx")    

            # apply missing selection
            msk = df[sample][year][ch]["fj_mass"]>40
            
            # apply tagger selection
            msk2 = (df[sample][year][ch]["THWW"]>0.905)
            df[sample][ch] = df[sample][year][ch][msk & msk2]

            # get nominal weight
            pkl_files = glob.glob(f"{samples_dir}_{year}/{samples_dict[sample]}/outfiles/*.pkl")

            df[sample][year][ch]["xsecweight"] = get_xsecweight(pkl_files, year, samples_dict[sample], False, luminosity)        
            df[sample][year][ch]["nominal"] = df[sample][year][ch]["xsecweight"] * df[sample][year][ch][f"weight_{ch}"]
            
```

```python
# combine channels
for sample in samples:
    l = []
    for year in years:
        for ch in channels:
            l += [df[sample][year][ch]]

    df[sample]["lep"] = pd.concat(l)
    
# combine signal
l = []
for sample in samples:
    l += [df[sample]["lep"]]

df["sig"] = pd.concat(l)
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 8))

# Create the 2D histogram
h = ax.hist2d(
    df["sig"]["recobin"],
    df["sig"]["STXS_finecat_simplified"],
    bins=(4, range(len(yticks_simplified) + 1)),
    cmap="Reds",
    weights=df["sig"]["nominal"],
)

# Extract the histogram values (h[0]) and normalize them along the rows (y-axis)
hist_values = h[0]
row_sums = np.sum(hist_values, axis=1)  # Sum of each row
row_sums[row_sums == 0] = 1  # Avoid division by zero
hist_values /= row_sums[:, np.newaxis]  # Normalize each row by its sum

# Replot the normalized histogram using the modified histogram values
h = ax.hist2d(
    df["sig"]["recobin"],
    df["sig"]["STXS_finecat_simplified"],
    bins=(4, range(len(yticks_simplified) + 1)),
    cmap="Reds",
    weights=df["sig"]["nominal"],
    density=False  # Set density to False as we already normalized manually
)

ax.set_xlim(0, 3)
cell_length = 3 / 4
bin_centers = [cell_length / 2, 2 * cell_length - cell_length / 2, 3 * cell_length - cell_length / 2, 4 * cell_length - cell_length / 2]

ax.set_xticks(bin_centers)
ax.set_xticklabels(["VBFcat", f"ggF [250, {300+shift}]", f"ggF [{300+shift}, {450+shift}]", f"ggF [{450+shift}, Inf]"], rotation=90)

# Add a colorbar
cbar = fig.colorbar(h[3], ax=ax)
cbar.set_label(f"ggF+VBF MC events", labelpad=20, fontsize=20)

ytick_positions = [i + 0.5 for i in range(len(yticks_simplified))]  # Center of each bin
ax.set_yticks(ytick_positions)
ax.set_yticklabels(yticks_simplified, fontsize=20)

# Annotate the values on the plot with normalized values
for i in range(len(hist_values)):
    for j in range(len(hist_values[i])):
        value = hist_values[i, j]
        ax.text(bin_centers[i], ytick_positions[j], f'{value:.2f}', ha='center', va='center', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```
